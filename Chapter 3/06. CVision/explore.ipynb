```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
bgr_image = cv2.imread('images/mms.png')

rgb_image = cv2.cvtColor(bgr_image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15,15))
plt.imshow(rgb_image)
cv2.imshow('image',bgr_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)

```

<img src='hsv_map.png'/>

```python
hsv_img = cv2.cvtColor(bgr_image,cv2.COLOR_BGR2HSV)
lower_b = (90,0,150)
upper_b = (140,255,255)

lower_g = (35,0,150)
upper_g = (85,255,255)

lower_r = (140,0,50)
upper_r = (180,255,255)
```

```python
mask_blue = cv2.inRange(hsv_img,lower_b,upper_b)
mask_green = cv2.inRange(hsv_img,lower_g,upper_g)
mask_red = cv2.inRange(hsv_img,lower_r,upper_r)
fig,ax = plt.subplots(4,figsize=(15,60))
blue = rgb_image.copy()
blue[mask_blue==0] = [0,0,0]
ax[0].imshow(blue)
ax[0].set_title('blue masked')

green = rgb_image.copy()
green[mask_green==0] = [0,0,0]
ax[1].imshow(green)
ax[1].set_title('green masked')

red = rgb_image.copy()
red[mask_red==0] = [0,0,0]
ax[2].imshow(red)
ax[2].set_title('red masked')

trio = rgb_image.copy()
trio[(mask_red==0)& (mask_blue==0) &(mask_green==0)] = [178,178,178]
ax[3].imshow(trio)
ax[3].set_title('red & green & blue masked')
```
