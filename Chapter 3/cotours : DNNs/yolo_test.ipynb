```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from yolo_helpers import *
```

## import config and weight files


```python
model_config = 'models/yolov3.cfg'
model_weights = 'models/yolov3.weights'
```

## loading the model

```python
net = cv2.dnn.readNetFromDarknet(model_config,model_weights)
```

## image->blob

```python
img = cv2.imread('images/person.jpg')
blob = cv2.dnn.blobFromImage(img,scalefactor=1/255.0,size=(416,416))
```

## setting input for model

```python
net.setInput(blob)
output_layer = getOutputsNames(net)
pred = net.forward(output_layer)
```

## cleaning prediction (bboxes)

```python
postprocess(img,pred)
```
