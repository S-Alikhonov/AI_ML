```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

```

```python
url ='https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data'
columns = ['buying','maint','doors','persons','lug_boot','safety','class']
df = pd.read_csv(url,header=None,names=columns)
df
```

```python
encode = OrdinalEncoder()
df[['buying','maint','lug_boot','safety']] = encode.fit_transform(df[['buying','maint','lug_boot','safety']].values)
df.doors[df.doors=='5more'] = '5'
df.persons[df.persons == 'more'] = '5'
df.iloc[:,:-1] = encode.fit_transform(df.iloc[:,:-1].values)
df
```

```python
X = df.iloc[:,:-1].values
y = df['class']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
clf = RandomForestClassifier(n_estimators=200,random_state=42,max_features=3)
```

```python
clf.fit(X_train,y_train)
pred = clf.predict(X_test)
```

```python
sum(pred==y_test)/pred.shape[0]
```

**max_features** can affect on prediction accuracy. If we use all features in each and every tree of the forest, they would get the same decision structure. instead, we should use random combination of features for trees. square root or logarithm of number features is suggested as the better value for max_features.
