```python
# continuos  num data augmentation

```

```python
age_std = np.std(df.age)
trbps_std = np.std(df.trtbps)
chol_std = np.std(df.chol)
thalachh_std = np.std(df.thalachh)
oldpeak_std = np.std(df.oldpeak)
data =[]
for _,row in df.sample(70).iterrows():
    temp = {
        'age':row['age']-np.random.uniform(age_std/5),
        'sex':row['sex'],
        'cp':row['cp'],
        'trtbps':row['trtbps']-np.random.uniform(trbps_std/5),
        'chol':row['chol']-np.random.uniform(chol_std/5),
        'fbs':row['fbs'],
        'restecg':row['restecg'],
        'thalachh':row['thalachh']-np.random.uniform(thalachh_std/5),
        'exng':row['exng'],
        'oldpeak':row['oldpeak']-np.random.uniform(oldpeak_std/5),
        'slp':row['slp'],
        'caa':row['caa'],
        'thall':row['thall'],
        'output':row['output']
        
    }
    data.append(temp)
    
    
```

```python
new= pd.DataFrame(data)
```

```python
stretched = df.append(new)
```

```python
stretched.to_csv('expanded.csv')
```
